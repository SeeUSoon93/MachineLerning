## 데이터 셋 구성하기
- 문제와 답 분리
- 훈련세트와 평가세트로 분리
    - 훈련세트/훈련데이터/train : 머신러닝 모델을 학습할 때 사용
    - 평가세트/평가데이터/홀드아웃 세트 : 모델이 얼마나 잘 작동하는지 측정하는데 사용

```python
import pandas as pd
```

```python
# 문제데이터 생성 -> 데이터프레임 형태로 생성
# columns 값은 feature_names 활용
X = pd.DataFrame(iris_data['data'], columns=iris_data['feature_names'])
X
```

```python
# 문제와 답 데이터로 분리
# 문제데이터는 대문자 X에 담기
# 답데이터는 소문자 y에 담기
y = pd.Series(iris_data['target'])
y
```

```python
# 훈련세트와 평가세트로 분리하기
# train_test_split 함수 : 훈련세트와 테스트세트를 분리해주는 기능
# 데이터를 나누기 전에 데이터셋을 무작위로 섞어준다.
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,
                                                    test_size=0.3,
                                                    random_state = 2)
```

```python
print("X_train 크기:", X_train.shape)
print("X_test 크기:", X_test.shape)
print("y_train 크기:", y_train.shape)
print("y_test 크기:", y_test.shape)
```

### EDA(데이터 살펴보기)

```python
# pd.plotting.scatter_matrix 함수 : 데이터프레임 산점도 행렬 생성
pd.plotting.scatter_matrix(X,
                            figsize=(15, 15), #행렬의 크기
                            marker='O',
                            c= y, # 그래프 점의 색상 / y는 클래스레이블
                            alpha=.5) # alpha는 점의 투명도 1로 갈수록 불투명
```

- 세 클래스가 꽃잎과 꽃받침의 측정값에 따라 비교적 잘 구분됨
- 머신러닝모델을 학습시키는데 적합하다고 판단됨

### 모델링

```python
# 모델 생성
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=2)
```

```python
# 모델 학습
knn.fit(X_train, y_train)
```

```python
# 예측하기
pre = knn.predict(X_test)
pre
```

```python
# 평가하기
from sklearn import metrics #평가를 위한 모듈
metrics.accuracy_score(pre,y_test)
```

```python
print("테스트 세트의 정확도: {:.2f}".format(metrics.accuracy_score(pre,y_test)))
```

## 하이퍼파리미터 튜닝

```python
knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(X_train, y_train)
pre = knn.predict(X_test)
metrics.accuracy_score(pre,y_test)
print("테스트 세트의 정확도: {:.2f}".format(metrics.accuracy_score(pre,y_test)))
```

```python
test_list = []
train_list = []

for k in range(1,105):
    #모델 생성
    knn = KNeighborsClassifier(n_neighbors=k)
    
    #모델 학습
    knn.fit(X_train, y_train)
    
    # test_list에 결과값을 저장
    test_pre = knn.predict(X_test)
    test_list.append(metrics.accuracy_score(test_pre,y_test))
    
    
    # train예측
    train_pre = knn.predict(X_train)
    # train 평가
    train_result = metrics.accuracy_score(train_pre,y_train)
    # 평가값을 train_list 결과값을 저장
    train_list.append(train_result)
```

```python
for i in range(len(test_list)):
    print("n_neighbors =",i+1,":",test_list[i],"/",train_list[i])
```

```python
# 훈련용/테스트 결과 시각화하기
import matplotlib.pyplot as plt

plt.figure(figsize=(10,10))
plt.plot(range(1,105), train_list, label='train')
plt.plot(range(1,105), test_list, label='test')

plt.xlabel('n_neighbors')
plt.ylabel('accuracy_scores', rotation=90)
plt.legend()
plt.show()
```

## sklearn 라이브러리의 k-NN 클래스 사용시 발생하는 특수한 에러 해결 방법
- sklearn 라이브러리의 KNeighborsClassifier() 를 사용하여 만든 모델 객체를 기반으로 예측값 산출을 위해 .predict() 메서드를 사용하는 경우 다음과 같은 에러가 발생하며 코드 실행이 되지 않는다.
- 해당 에러는 sklearn 1.3.0 버전 때문에 발생한 문제로 다른 버전으로 재설치를 하면 해결이 될 수 있다. 대표적인 해결 방법은 버전을 1.2.2로 다운그레이드 하는 것이며 주피터노트북 환경에서 다음의 코드를 실행하면 된다.
